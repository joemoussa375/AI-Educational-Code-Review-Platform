## Core Classes (with Performance Instrumentation)

```python
# ============================================
# KNOWLEDGE BASE with Timing
# ============================================

class KnowledgeBase:
    def __init__(self, url="https://peps.python.org/pep-0008/"):
        start_time = time.time()
        print(f"[INFO] Ingesting Knowledge from: {url}")
        
        try:
            loader = WebBaseLoader(url)
            data = loader.load()
        except Exception as e:
            print(f"[WARNING] Using backup rules: {e}")
            backup_text = """
            Function Names: Function names should be lowercase, with words separated by underscores.
            Class Names: Class names should normally use the CapWords convention.
            Imports: Imports should usually be on separate lines and at the top of the file.
            Whitespace: Avoid extraneous whitespace immediately inside parentheses.
            Constants: Constants should be defined in all capital letters.
            """
            data = [Document(page_content=backup_text)]

        text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
        splits = text_splitter.split_documents(data)
        
        embedding_model = HuggingFaceEmbeddings(
            model_name="all-MiniLM-L6-v2",
            model_kwargs={'device': 'cpu'}
        )

        self.vectorstore = FAISS.from_documents(splits, embedding_model)
        self.bm25_retriever = BM25Retriever.from_documents(splits)
        self.bm25_retriever.k = 3
        
        self.init_time = time.time() - start_time
        print(f"[OK] Knowledge Base Ready. Init time: {self.init_time:.2f}s")

    def search(self, queries):
        start_time = time.time()
        results = []
        for q in queries:
            docs_faiss = self.vectorstore.similarity_search(q, k=2)
            docs_bm25 = self.bm25_retriever.invoke(q)
            results.extend(docs_faiss + docs_bm25)
        
        unique_docs = {doc.page_content: doc for doc in results}.values()
        search_time = time.time() - start_time
        return "\n\n".join([f"[Guide]: {doc.page_content}" for doc in unique_docs]), search_time
```

```python
# ============================================
# UNIFIED CODE REVIEWER with Metrics Collection
# ============================================

class UnifiedCodeReviewer:
    def __init__(self, model_id="codellama/CodeLlama-7b-Instruct-hf"):
        self.metrics = {}  # Store timing metrics
        
        # Initialize RAG
        self.kb = KnowledgeBase()
        
        print(f"[INFO] Loading Model: {model_id}...")
        start_time = time.time()
        
        bnb_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_compute_dtype=torch.float16,
            bnb_4bit_use_double_quant=True,
        )

        self.tokenizer = AutoTokenizer.from_pretrained(model_id)
        self.model = AutoModelForCausalLM.from_pretrained(
            model_id,
            quantization_config=bnb_config,
            device_map="auto"
        )
        
        self.model_load_time = time.time() - start_time
        print(f"[OK] Model loaded in {self.model_load_time:.2f}s")
        
        # Get VRAM usage after model load
        if torch.cuda.is_available():
            self.vram_allocated = torch.cuda.memory_allocated() / (1024**3)
            self.vram_reserved = torch.cuda.memory_reserved() / (1024**3)
            print(f"[INFO] VRAM Allocated: {self.vram_allocated:.2f} GB")
            print(f"[INFO] VRAM Reserved: {self.vram_reserved:.2f} GB")

    def _ask_llm(self, prompt, max_tokens=200):
        start_time = time.time()
        inputs = self.tokenizer(prompt, return_tensors="pt").to("cuda")
        outputs = self.model.generate(
            **inputs, 
            max_new_tokens=max_tokens,
            temperature=0.1
        )
        result = self.tokenizer.decode(outputs[0], skip_special_tokens=True).split("[/INST]")[-1].strip()
        return result, time.time() - start_time

    def _run_pylint(self, code_snippet):
        start_time = time.time()
        with tempfile.NamedTemporaryFile(mode='w+', suffix='.py', delete=False) as f:
            f.write(code_snippet)
            fname = f.name
            
        try:
            result = subprocess.run(
                ["pylint", fname, "--output-format=json"], 
                capture_output=True, text=True
            )
            
            if not result.stdout.strip(): 
                return "No Syntax Errors Found.", time.time() - start_time
            
            try:
                errors = json.loads(result.stdout)
                if len(errors) > 15:
                    report = [f"Line {e['line']}: {e['message']} ({e['symbol']})" for e in errors[:15]]
                    report.append("... (and more)")
                else:
                    report = [f"Line {e['line']}: {e['message']} ({e['symbol']})" for e in errors]
                return "\n".join(report), time.time() - start_time
            except json.JSONDecodeError: 
                return "Pylint ran but output was not parseable.", time.time() - start_time
        except Exception as e:
            return f"Pylint Failed to Run: {e}", time.time() - start_time
        finally:
            if os.path.exists(fname): 
                os.remove(fname)

    def _generate_search_plan(self, code_snippet):
        prompt = f"""<s>[INST] <<SYS>>
        You are a Technical Analyst.
        Identify 3 SPECIFIC Python Style topics relevant to the code provided.
        Output valid JSON only.
        {{ "queries": ["topic 1", "topic 2", "topic 3"] }}
        <</SYS>>
        Code: {code_snippet}
        JSON: [/INST]"""
        
        raw_response, plan_time = self._ask_llm(prompt, max_tokens=150)
        
        try:
            json_match = re.search(r'\{.*\}', raw_response, re.DOTALL)
            if json_match:
                data = json.loads(json_match.group(0))
                return data.get("queries", []), plan_time
            else:
                return [raw_response.strip()], plan_time
        except json.JSONDecodeError:
            return ["General Python Style"], plan_time

    def review(self, code_snippet):
        """
        Main review function with detailed metrics collection.
        Returns: (result, metrics_dict)
        """
        total_start = time.time()
        metrics = {}
        
        # Phase 1: Static Analysis
        print("  1. Running Pylint...")
        linter_report, metrics['pylint_time'] = self._run_pylint(code_snippet)
        
        # Phase 2: Search Planning
        print("  2. Generating search plan...")
        search_terms, metrics['planning_time'] = self._generate_search_plan(code_snippet)
        metrics['search_terms'] = search_terms
        
        # Phase 3: RAG Retrieval
        print("  3. Retrieving style rules...")
        rag_rules, metrics['rag_time'] = self.kb.search(search_terms)
        
        # Phase 4: Final Analysis
        print("  4. Generating final report...")
        final_prompt = f"""<s>[INST] <<SYS>>
        You are a Senior Software Architect and Python Expert.
        Conduct a comprehensive code review using the provided reports.
        
        INPUTS:
        1. STATIC ANALYSIS REPORT:
        {linter_report}
        
        2. STYLE GUIDE RULES:
        {rag_rules}
        
        PRINCIPLES: 
        1. Immutability & Safety: Prefer creating new data structures over mutating existing ones.
        2. Resource Management: Ensure all resources are properly managed.
        3. Algorithmic Efficiency: Identify unnecessary complexity.
        4. Style Compliance: Verify naming and formatting.
        
        OUTPUT FORMAT:
        **1. Critical Issues:** (Logical bugs, Safety risks)
        **2. Style Analysis:** (PEP 8 compliance)
        **3. Refactored Solution:** (Corrected code)
        <</SYS>>

        User Code:
        {code_snippet} [/INST]"""
        
        result, metrics['llm_time'] = self._ask_llm(final_prompt, max_tokens=1024)
        
        # Record totals
        metrics['total_time'] = time.time() - total_start
        metrics['pylint_report'] = linter_report
        
        # Get VRAM after inference
        if torch.cuda.is_available():
            metrics['vram_peak'] = torch.cuda.max_memory_allocated() / (1024**3)
        
        # Cleanup
        gc.collect()
        torch.cuda.empty_cache()
        
        return result, metrics
```

## 5.3 Performance Metrics - Full Benchmark

```python
# ============================================
# FULL BENCHMARK: Run all 10 test cases
# ============================================
print("=" * 70)
print("FULL BENCHMARK: Running all 10 test cases")
print("=" * 70)

all_results = {}

for test_name, test_data in TEST_CASES.items():
    print(f"\nRunning: {test_name}...")
    
    try:
        result, metrics = engine.review(test_data['code'])
        all_results[test_name] = {
            'description': test_data['description'],
            'category': test_data['category'],
            'expected_issues': test_data['expected_issues'],
            'result': result,
            'metrics': metrics
        }
        print(f"  Completed in {metrics['total_time']:.2f}s")
    except Exception as e:
        print(f"  Error: {e}")
        all_results[test_name] = {'error': str(e)}

print("\n" + "=" * 70)
print("BENCHMARK COMPLETE")
print("=" * 70)
```

```python
# ============================================
# PERFORMANCE SUMMARY TABLE
# ============================================
print("\n" + "=" * 90)
print("PERFORMANCE METRICS SUMMARY")
print("=" * 90)
print(f"{'Test Case':<25} {'Category':<12} {'Pylint':<10} {'RAG':<10} {'LLM':<10} {'Total':<10}")
print("-" * 90)

total_times = []
pylint_times = []
rag_times = []
llm_times = []

for test_name, data in all_results.items():
    if 'error' not in data:
        m = data['metrics']
        print(f"{test_name:<25} {data['category']:<12} {m['pylint_time']:.3f}s    {m['rag_time']:.3f}s    {m['llm_time']:.2f}s    {m['total_time']:.2f}s")
        total_times.append(m['total_time'])
        pylint_times.append(m['pylint_time'])
        rag_times.append(m['rag_time'])
        llm_times.append(m['llm_time'])

print("-" * 90)
print(f"{'AVERAGE':<25} {'---':<12} {sum(pylint_times)/len(pylint_times):.3f}s    {sum(rag_times)/len(rag_times):.3f}s    {sum(llm_times)/len(llm_times):.2f}s    {sum(total_times)/len(total_times):.2f}s")
print(f"{'MIN':<25} {'---':<12} {min(pylint_times):.3f}s    {min(rag_times):.3f}s    {min(llm_times):.2f}s    {min(total_times):.2f}s")
print(f"{'MAX':<25} {'---':<12} {max(pylint_times):.3f}s    {max(rag_times):.3f}s    {max(llm_times):.2f}s    {max(total_times):.2f}s")
print("=" * 90)
```
